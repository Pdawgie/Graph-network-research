# ER

```python
class Graph_Monte_Carlo_erdos_renyi_graph():

    def sim_katz(self,n):
        ls=[]
        for i in range(n):
            p=0.08645498148260579
            g=nx.fast_gnp_random_graph(182,p)
            katz=dict(nx.katz_centrality(g,tol=1.5)).values()
            ave_katz=sum(katz)/len(katz)
            ls.append(ave_katz)
        return ls

    def sim_average_clustering(self,n):
        ls=[]
        for i in range(n):
            p=0.08645498148260579
            g=nx.fast_gnp_random_graph(182,p)
            a_clus = nx.average_clustering(g)
            ls.append(a_clus)
        return ls


    def sim_degenracy(self,n):
        ls=[]
        for i in range(n):
            p=0.08645498148260579
            g=nx.fast_gnp_random_graph(182,p)
            degen=max(nx.core_number(g).values())
            ls.append(degen)
        return ls

    def node_connectivity(self,n): #comp ineff
        ls=[]
        for i in range(n):
            p=0.08645498148260579
            g=nx.fast_gnp_random_graph(182,p)
            conn=nx.node_connectivity(g)
            ls.append(conn)
        return ls

    def sim_aggregate_constraint(self,n):
        ls=[]
        for i in range(n):
            p=0.08645498148260579
            g=nx.fast_gnp_random_graph(182,p)
            constraint=nx.betweenness_centrality(g).values() #brokerage and aggregate constraint
            avg_constraint = [v for v in constraint]
            agg=sum(avg_constraint)/len(avg_constraint)
            ls.append(agg)
        return ls
    
    def sim_transivity(self,n):
        ls=[]
        for i in range(n):
            p=0.08645498148260579
            g=nx.fast_gnp_random_graph(182,p)
            trans=nx.transitivity(g)
            ls.append(trans)
        return ls
        
    def sim_density(self,n):
        ls=[]
        for i in range(n):
            p=0.08645498148260579
            g=nx.fast_gnp_random_graph(182,p)
            dens=nx.density(g)
            ls.append(dens)
        return ls

    def sim_average_degree_conn(self,n): #  Compute the average shortest path length for each node
        ls=[]
        for i in range(n):
            p=0.08645498148260579
            g=nx.fast_gnp_random_graph(182,p)
            degree_conn=nx.average_degree_connectivity(g).values()
            degcon=sum(degree_conn)/len(degree_conn)
            ls.append(degcon)
        return ls

    def sim_avg_spl(self,n):
        ls=[]
        for i in range(n):
            p=0.08645498148260579
            g=nx.fast_gnp_random_graph(182,p)
            shortest_path_lengths = dict(nx.all_pairs_shortest_path_length(g))
            av_len=np.mean([np.mean(list(spl.values())) for spl in shortest_path_lengths.values()])
            ls.append(av_len)
        return ls

    def sim_graph_betweeness(self,n):
        ls=[]
        for i in range(n):
            p=0.08645498148260579
            g=nx.fast_gnp_random_graph(182,p)
            bet_centrality = nx.betweenness_centrality(g).values()
            max_bet_cen=max(bet_centrality)
            diff=[max_bet_cen-x for x in bet_centrality]
            graph_bet_cen=sum(diff)/max(diff)
            ls.append(graph_bet_cen)
        return ls


    def sim_diameter(self,n):
        ls=[]
        for i in range(n):
            p=0.08645498148260579
            g=nx.fast_gnp_random_graph(182,p)
            diameter=max([max(j.values()) for (i,j) in nx.shortest_path_length(g)])
            ls.append(diameter)
        return ls

    def sim_radius(self,n):
        ls=[]
        for i in range(n):
            p=0.08645498148260579
            g=nx.fast_gnp_random_graph(182,p)
            radius_dat=[max(j.values()) for (i,j) in nx.shortest_path_length(g)]
            rad=min([i for i in radius_dat if i != 0])
            ls.append(rad)
        return ls


    def sim_assortativity(self,n):
        ls=[]
        for i in range(n):
            p=0.08645498148260579
            g=nx.fast_gnp_random_graph(182,p)
            assort=nx.degree_pearson_correlation_coefficient(g)
            ls.append(assort)
        return ls


    def sim_algebraic_conn(self,n):
        ls=[]
        for i in range(n):
            p=0.08645498148260579
            g=nx.fast_gnp_random_graph(182,p)
            L_di=nx.laplacian_matrix(g)
            el = np.linalg.eigvals(L_di.A)
            algebraic_conn=sorted(el)[::-1][1]
            ls.append(algebraic_conn)
        return ls


    def sim_laplacian_spectral_rad(self,n):
        ls=[]
        for i in range(n):
            p=0.08645498148260579
            g=nx.fast_gnp_random_graph(182,p)
            L_di=nx.laplacian_matrix(g)
            el = np.linalg.eigvals(L_di.A)
            Laplacian_spectral_rad=max(el)
            ls.append(Laplacian_spectral_rad)
        return ls


    def sim_network_efficiency(self,n):
        def all_shortest_paths(G):
            a = list(nx.all_pairs_shortest_path(G))
            all_sp_list = []
            for n in range(len(G.nodes)):
                a1 = a[n][1]
                for k,v in a1.items():
                    all_sp_list.append(len(v))
            return all_sp_list
        ls=[]
        for i in range(n):
            p=0.08645498148260579
            g=nx.fast_gnp_random_graph(182,p)
            N=g.number_of_nodes()
            efficiency=(1/(N*(N-1)))*sum(np.reciprocal(all_shortest_paths(g)))
            ls.append(efficiency)

        return ls

    def sim_spectral_rad(self,n):
        ls=[]
        for i in range(n):
            p=0.08645498148260579
            g=nx.fast_gnp_random_graph(182,p)
            adj=nx.adjacency_matrix(g)
            spec_rad = max(np.linalg.eigvals(adj.A))
            ls.append(spec_rad)
        print(np.mean(ls))
        return ls

    def sim_entropy(self,n):
        ls=[]
        for i in range(n):
            p=0.08645498148260579
            g=nx.fast_gnp_random_graph(182,p)
            degs=dict(g.degree()).values()
            N=nx.number_of_nodes(g)
            pij=[np.log(x) for x in degs]
            Shannon_ent=(1/(N*np.log(N-1)))*sum(pij)
            ls.append(Shannon_ent)
        print(np.mean(ls))
        return ls
```
